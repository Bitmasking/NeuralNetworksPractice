```python
dataArray = np.loadtxt("Data.txt")
[rows, cols] = dataArray.shape

theta = [2, 5, 7]
n_samples = rows

y = dataArray[:,2]
x0 = np.ones((n_samples, 1))
x1 = np.reshape(dataArray[:,0], (n_samples, 1))
x2 = np.reshape(dataArray[:,1], (n_samples, 1))
X = np.hstack((x0, x1, x2))

y = X @ theta
plt.rcParams["figure.figsize"] = [16,9]
plt.rc('axes',edgecolor='green')

ax = plt.subplot(projection='3d')
ax.plot(X[:,1], X[:,2], y, '.')

ax.set(
    xlabel='$x_1$',
    ylabel='$x_2$',
    zlabel='y'
)
ax.view_init(15, 85)
plt.tight_layout()
```

```python
def ordinary_least_squares(X, y):
  theta_hat = np.linalg.inv(X.T @ X) @ X.T @ y

  return theta_hat
theta_hat = ordinary_least_squares(X, y)
print(theta_hat)
```

```python
# Prediccion
theta_hat = ordinary_least_squares(X, y)
y_hat = X @ theta_hat

# Computamos el MSE
print(f"MSE = {np.mean((y - y_hat)**2):.2f}")
```
